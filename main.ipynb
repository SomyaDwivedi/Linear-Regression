```python
x = [153, 200, 140, 210, 170]
y = [900, 923, 880, 930, 910]
```

```python
import matplotlib.pyplot as plt
plt.scatter(x, y)
```

y = w1(X) + w0

w1 =  ( summation (Xi - Xmean)(yi - Ymean) ) / summation (Xi - Xmean)^2

w0 = ymean  - w1(Xmean)

```python
Xmean = sum(x) / len(x)
Ymean = sum(y) / len(y)

a = 0
b = 0
for i in range(len(x)):
    a += (x[i] - Xmean) * (y[i] - Ymean)
    b += (x[i] - Xmean) ** 2

w1 = a / b
w0 = Ymean - (w1 * Xmean)

x = float(input("Enter Price (X): "))
y = (w1*x )+ w0
print(y)    
```

```python
import numpy as np

plt.scatter([153, 200, 140, 210, 170], [900, 923, 880, 930, 910], color='blue', label='Data Points')


x_values = np.array([min([153, 200, 140, 210, 170]), max([153, 200, 140, 210, 170])])
y_values = w1 * x_values + w0
plt.plot(x_values, y_values, color='red', label='Regression Line')

plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```
