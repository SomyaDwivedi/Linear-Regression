```python
import matplotlib.pyplot as plt
import numpy as np
import math

x = [140, 153, 170, 200, 210, 180, 160, 220, 230, 250, 240, 160, 180, 200, 190, 205, 215, 225, 235, 245]
y = [880, 900, 910, 923, 930, 915, 905, 935, 940, 950, 945, 910, 920, 925, 915, 930, 935, 940, 950, 960]
```

```python
# Calculate means
Xmean = sum(x) / len(x)
Ymean = sum(y) / len(y)

# Calculate w1 and w0 (linear regression coefficients)
a = 0
b = 0
for i in range(len(x)):
    a += (x[i] - Xmean) * (y[i] - Ymean)
    b += (x[i] - Xmean) ** 2

w1 = a / b
w0 = Ymean - (w1 * Xmean)
```

```python
# Predict y using the regression model
X_input = int(input("Enter Price (X): "))
Y_pred = (w1 * X_input) + w0
print(f"Predicted Y for price {X_input} is: {Y_pred}")
```

```python
# Plotting
plt.scatter(x, y, color='blue', label='Data Points')

x_values = np.array([min(x), max(x)])
y_values = w1 * x_values + w0
plt.plot(x_values, y_values, color='red', label='Regression Line')

plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

```python
# R² Score calculation
sse = 0.0
sst = 0.0

for i in range(len(x)):
    y_pred = (w1 * x[i]) + w0
    sse += (y[i] - y_pred)**2
    sst += (y[i] - Ymean)**2

r2_score = 1 - (sse / sst)

print(f"The model's R² score is {r2_score * 100}% accurate")
```

```python
# Mean Squared Error (MSE)
mse = sse / len(x)
print(f"The Mean Squared Error is {mse}, which means the answer can be off by ±{math.sqrt(mse)} units")
```
